# Eval

```python
path_models = list(Path(config.data.training.save_dir).glob('*.pt'))

accuracy , auc, cm, recall, precision, f1, confusion_matrix = [], [], [], [], [], [], []
epochs = []

for path_model in path_models: 
    
    print(path_model)
    saved_model = torch.load(path_model)
    accuracy.append(saved_model['acc'])
    auc.append(saved_model['auc'])
    recall.append(saved_model['recall'])
    precision.append(saved_model['precision'])
    f1.append(saved_model['f1'])
    confusion_matrix.append(saved_model['confusion_matrix'])
    epochs.append(saved_model['epoch'])
    #cm.append(saved_model['confusion_matrix'])
```
